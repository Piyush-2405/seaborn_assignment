Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a powerful Python data visualization library based on Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics. Here are five commonly used plots in Seaborn and their uses:

1.Scatter Plot:
Use: Scatter plots display individual data points as points on a 2D plane, with one variable on the x-axis and another on the y-axis. They are useful for visualizing the relationship between two continuous variables and identifying patterns or trends.

2.Bar Plot:
Use: Bar plots represent data using rectangular bars, with the length or height of each bar proportional to the values they represent. They are suitable for displaying and comparing discrete data, such as categorical variables and their associated counts or aggregated values.

3.Histogram:
Use: Histograms are used to visualize the distribution of a single numerical variable. The data is divided into bins, and the height of each bar represents the frequency or count of data points within that bin.

4.Box Plot (Box-and-Whisker Plot):
Use: Box plots are used to visualize the distribution of a numerical variable across different categories or groups. They show the median, quartiles, and potential outliers, making them useful for identifying the spread and central tendency of the data.

5.Heatmap:
Use: Heatmaps are graphical representations of data using colors to visualize a matrix of values. They are particularly useful for displaying correlations between variables or to visualize the patterns in large datasets.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
fmri = sns.load_dataset('fmri')
```

```python
fmri
```

```python
import matplotlib.pyplot as plt

# Create the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
plt.title("Line Plot of 'timepoint' vs. 'signal' for different events and regions")
#plt.xlabel("Timepoint")
#plt.ylabel("Signal")
plt.legend(title="Event", loc="best")
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights_dataset = sns.load_dataset('flights')
```

```python
flights_dataset
```

```python
flights_pivot = flights_dataset.pivot("month", "year", "passengers")
```
